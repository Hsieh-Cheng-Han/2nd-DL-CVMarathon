```python
import cv2
import matplotlib.pyplot as plt
import time
import numpy as np

%matplotlib inline

# image path
img_path = 'lena.png'

# read the image in full color
img_color = cv2.imread(img_path, cv2.IMREAD_COLOR)
img_color = cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB)

# show image in full color
plt.imshow(img_color)
```

# Horizontal and Vertical

```python
# horizontal
img_hflip = img_color[::-1,:,:]
print('Horizontal reverse:')
plt.imshow(img_hflip)
plt.show()

# vertical
img_vflip = img_color[:,::-1,:]
print('Vertical reverse:')
plt.imshow(img_vflip)
plt.show()

# horizontal + vertical
img_hvflip = img_color[::-1,::-1,:]
print('Horizontal and vertical reverse:')
plt.imshow(img_hvflip)
plt.show()
```

# Contraction and Enlargement

```python
# Contraction
img_test = cv2.resize(img_color, None, fx = 0.2, fy = 0.2)

# Enlargement for 8 times
fx, fy = 8, 8

# Inter_Nearest interpolation
start_time = time.time()
img_area_scale = cv2.resize(img_test, None, fx = fx, fy = fy, interpolation = cv2.INTER_NEAREST)
print('INTER_NEAREST zoom cost {}'.format(time.time() - start_time))
plt.imshow(img_area_scale)
plt.show()

# Inter_Nearest interpolation
start_time = time.time()
img_area_scale = cv2.resize(img_test, None, fx = fx, fy = fy, interpolation = cv2.INTER_CUBIC)
print('INTER_CUBIC zoom cost {}'.format(time.time() - start_time))
plt.imshow(img_area_scale)
plt.show()
```

Inter_cubic method is slower but has better quality reconstrution.

# Translation Transformation

```python
# set up translation transformation matrix
# x : 50 pixel; y : 100 pixel
M = np.array([[1, 0, 50],
              [0, 1, 100]], dtype = np.float32)
shift_img = cv2.warpAffine(img_color, M, (img_color.shape[1], img_color.shape[0]))
print('After translation transformation:')
plt.imshow(shift_img)
plt.show()
```
