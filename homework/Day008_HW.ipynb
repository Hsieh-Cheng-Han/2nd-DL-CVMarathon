```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

# image path
img_path = 'lena.png'

# read the image in full color
img = cv2.imread(img_path, cv2.IMREAD_COLOR)

# show the image in RGB form:
def show_image_RGB(img_BGR):
    plt.imshow(cv2.cvtColor(img_BGR, cv2.COLOR_BGR2RGB))
    plt.show()

print('Original picture :')
show_image_RGB(img)    
```

```python
# turn it to gray scale
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Sobel detection in unit16
img_sobel_x = cv2.Sobel(img_gray, cv2.CV_16S, dx=1, dy=0, ksize=3)

# convert scale
img_sobel_x = cv2.convertScaleAbs(img_sobel_x)

# Sobel detection in unit8
img_sobel_x_uint8 = cv2.Sobel(img_gray, cv2.CV_8U, dx=1, dy=0, ksize=3)
```

```python
print('Unit16 :')
plt.imshow(img_sobel_x, cmap = 'gray')
plt.show()
print('Unit8 :')
plt.imshow(img_sobel_x_uint8, cmap = 'gray')
plt.show()
```

```python
# one-time derivative
img_sobel_x = cv2.convertScaleAbs(cv2.Sobel(img_gray, cv2.CV_16S, dx=1, dy=0, ksize=3))

# two-time derivative
img_sobel_xx = cv2.convertScaleAbs(cv2.Sobel(img_gray, cv2.CV_16S, dx=2, dy=0, ksize=3))

print('one-time derivative :')
plt.imshow(img_sobel_x, cmap = 'gray')
plt.show()
print('wo-time derivative :')
plt.imshow(img_sobel_xx, cmap = 'gray')
plt.show()
```
