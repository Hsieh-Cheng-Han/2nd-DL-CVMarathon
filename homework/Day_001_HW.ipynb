```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

# image path
img_path = 'lena.png'

# read the image in full color
img_color = cv2.imread(img_path, cv2.IMREAD_COLOR)

# read the image in gray scale
img_gray = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

# show image in full color
plt.imshow(cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB))

# show image in gray scale
plt.imshow(img_gray, cmap = 'gray')
```

```python
# Check each channel's performance
# Since there is only one channel, it looks like a gray-scale image
for i in range(3):
    print('The '+str(i+1)+'th channel:')
    plt.imshow(cv2.cvtColor(img_color[:,:,i:i+1], cv2.COLOR_BGR2RGB))
    plt.show()
```
