```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

# image path
img_path = 'lena.png'

# read the image in full color
img_BGR = cv2.imread(img_path, cv2.IMREAD_COLOR)

# show the image in RGB form:
def show_image_RGB(img_BGR):
    plt.imshow(cv2.cvtColor(img_BGR, cv2.COLOR_BGR2RGB))
    plt.show()

print('Original picture :')
show_image_RGB(img_BGR)    
```

```python
# histogram equalization
img_BGR_new = img_BGR.copy()
img_HSV_new = cv2.cvtColor(img_BGR_new, cv2.COLOR_BGR2HSV)
img_HSV_new[:,:,2] = cv2.equalizeHist(img_HSV_new[:,:,2])
img_BGR_new = cv2.cvtColor(img_HSV_new, cv2.COLOR_HSV2BGR)

show_image_RGB(img_BGR_new)
```

```python
# vertical reverse and contraction
img_BGR_new = img_BGR_new[:,::-1,:]
img_BGR_new = cv2.resize(img_BGR_new, None, fx = 0.5, fy = 0.5, interpolation = cv2.INTER_LINEAR)
show_image_RGB(img_BGR_new)
```

```python
# Point transformation
point_1 = [60, 40]
point_2 = [420, 510]

original_weight = img_BGR.shape[0]
original_height = img_BGR.shape[1]

# vertical reverse
point_1_new = [original_weight - point_2[0], point_1[1]]
point_2_new = [original_weight - point_1[0], point_2[1]]

# contraction
point_1_new = [int(point*0.5) for point in point_1_new]
point_2_new = [int(point*0.5) for point in point_2_new]

print(point_1_new, point_2_new)
```

```python
# Draw the rectangular
cv2.rectangle(img_BGR_new, tuple(point_1_new), tuple(point_2_new), (0, 0, 255), 3)
show_image_RGB(img_BGR_new)
```
